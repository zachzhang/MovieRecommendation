```python
from imdb import IMDb
ia = IMDb()

import pandas as pd
import numpy as np
import csv
import urllib

!mkdir data/moviecovers

movieMap = pd.read_csv('ml-latest-small/movies.csv', nrows = 3)
movies = movieMap['title']


ia = IMDb()

all_movie_data = []
the_matrix = ia.get_movie('0133093')
#keys = the_matrix.keys()
keys = ['rating', 'runtimes', 'year', 'languages', 'votes', 'producer', 'title', 'writer', 'mpaa',
 'top 250 rank', 'editor', 'certificates', 'country codes', 'language codes', 'cover url', 
 'genres', 'director', u'production companies', 'countries', 'plot outline', 'plot', 'cast',
 'original music', 'full-size cover url']
#keys = ['cast', 'plot', 'director' ]
all_movie_data = [keys]
for midx, movie in enumerate(movies):
    
    movie_search_name = movie
    s_result = []
    while len(s_result) == 0 and len(movie_search_name) > 0:
        try:
            s_result = ia.search_movie(movie_search_name)
        except:
            try:
                s_result = ia.search_movie(movie_search_name)
            except:
                try:
                    s_result = ia.search_movie(movie_search_name)
                except:
                    s_result = ia.search_movie(movie_search_name)
        movie_search_name = movie_search_name.rsplit(' ', 1)[0]
    try:

        print movie

        s_result = ia.search_movie(movie)

        imdb_id = s_result[0].movieID

        imdb_movie = ia.get_movie(imdb_id)

        features=[]

        for key in keys:

            if  key in imdb_movie.keys():# and \
                #(isinstance(imdb_movie[key], str) or isinstance(imdb_movie[key], unicode)):
                if isinstance(imdb_movie[key], str):
                    features.append(imdb_movie[key].decode('unicode_escape').encode('ascii', 'ignore').strip())
                elif isinstance(imdb_movie[key], list) and isinstance(imdb_movie[key][0], str):
                    features.append(imdb_movie[key].decode('unicode_escape').encode('ascii', 'ignore').strip())
                elif isinstance(imdb_movie[key], list) and \
                    isinstance(imdb_movie[key][0], list) and isinstance(imdb_movie[key][0][0], str):
                    features.append(imdb_movie[key].decode('unicode_escape').encode('ascii', 'ignore').strip())
                else:
                    features.append(imdb_movie[key])
            else:
                features.append(np.nan)
                #if key in imdb_movie.keys():
                    #print(type(imdb_movie[key]))
            #print(features)
        features.append(movie)
        try:
            coverurl = imdb_movie['cover url']
            urllib.urlretrieve(coverurl, "data/moviecovers/" + movie)
        except:
            pass

        all_movie_data.append( features)

    except Exception as e:
        print(e)
        pass

with open('movieDataList.csv', 'wb') as myfile:
    wr = csv.writer(myfile)
    for movie_features in all_movie_data:
        try:
            wr.writerow(movie_features)
        except:
            pass
    
#all_movie_data_df = pd.DataFrame(all_movie_data ,columns = keys)
#all_movie_data_df['movie'] = movies

#all_movie_data_df.to_csv('movieDatasearchbroken2.csv')
```
